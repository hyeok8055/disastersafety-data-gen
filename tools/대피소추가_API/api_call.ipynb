```python
import requests
import urllib3
import pandas as pd
import numpy as np

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

url = "https://www.safetydata.go.kr/V2/api/DSSP-IF-00195"
serviceKey = "0YP09R0RZP9PV65C"
```

```python
# Initialize empty list to store all data
all_data = []

# Make API calls for pages 1-40
for page in range(1, 41):
    payloads = {
        "serviceKey": serviceKey,
        "returnType": "json",
        "pageNo": str(page),
        "numOfRows": "1000",
    }
    
    response = requests.get(url, params=payloads)
    
    # Check if response contains data
    if 'body' in response.json():
        all_data.extend(response.json()['body'])
```

```python
# Create DataFrame from all collected data
df_nationwide = pd.DataFrame(all_data)

print(f"전국 데이터 총 {len(df_nationwide)}개 수집 완료")
print("전국 데이터 컬럼:", df_nationwide.columns.tolist())
```

```python
# 대구광역시만 필터링
df_daegu = df_nationwide[df_nationwide['MNG_INST_NM'].str.contains("대구광역시", na=False)].copy()

print(f"대구광역시 데이터 총 {len(df_daegu)}개 필터링 완료")
print("대구 관리기관 목록:")
print(df_daegu['MNG_INST_NM'].value_counts())
```

대피소구분코드(한파쉼터:1,무더위쉼터:2,지진옥외대피장소:3,지진해일긴급대피장소:4)	
시설구분코드

```python
# 전국 데이터 처리
def process_shelter_data(df, data_name):
    """대피소 데이터 처리 함수"""
    df_processed = df.copy()
    
    # Convert degrees, minutes, seconds to decimal degrees
    df_processed['경도'] = df_processed['LOT_PROVIN'] + df_processed['LOT_MIN']/60 + df_processed['LOT_SEC']/3600
    df_processed['위도'] = df_processed['LAT_PROVIN'] + df_processed['LAT_MIN']/60 + df_processed['LAT_SEC']/3600

    # str 문자열 처리
    df_processed['FCLT_SE_CD'] = df_processed['FCLT_SE_CD'].str.strip()

    # Create mapping dictionary for facility codes
    facility_type_map = {
        '1': '한파',
        '2': '무더위', 
        '3': '지진',
        '4': '지진해일'
    }

    # Convert facility codes before renaming
    df_processed['FCLT_SE_CD'] = df_processed['FCLT_SE_CD'].map(facility_type_map)

    # Fill zero values in SHNT_PSBLTY_NOPE with FCLT_SCL values
    df_processed.loc[df_processed['SHNT_PSBLTY_NOPE'] == '0', 'SHNT_PSBLTY_NOPE'] = df_processed.loc[df_processed['SHNT_PSBLTY_NOPE'] == '0', 'FCLT_SCL']

    # Rename columns and select relevant ones
    result = df_processed.rename(columns={
        'FCLT_SE_CD': '시설구분코드',
        'FCLT_CD': '시설코드', 
        'FCLT_NM': '시설명',
        'FCLT_SCL': '시설규모',
        'SHNT_PSBLTY_NOPE': '대피가능인원수',
        'MNG_INST_NM': '관리기관명',
        'MNG_INST_TELNO': '관리기관전화번호',
        'FCLT_ADDR_RONA': '시설주소도로명',
        'OPN_YN': '개방여부',
        'GRND_UDGD_SE': '지상지하구분'
    })[['시설구분코드', '시설코드', '시설명', '시설규모', '대피가능인원수', '관리기관명', '관리기관전화번호',
         '시설주소도로명', '경도', '위도', '개방여부', '지상지하구분']]
    
    print(f"{data_name} 데이터 처리 완료: {len(result)}개")
    return result

# 전국 데이터 처리
processed_nationwide = process_shelter_data(df_nationwide, "전국")

print("\n=== 처리된 전국 데이터 샘플 ===")
processed_nationwide.head()
```

```python
# 대구 데이터 처리
processed_daegu = process_shelter_data(df_daegu, "대구광역시")

print("\n=== 처리된 대구 데이터 샘플 ===")
processed_daegu.head()
```

```python
# 전국 데이터 저장
processed_nationwide.to_csv('전국_대피소_데이터.csv', index=False, sep='|', encoding='utf-8-sig')
print(f"전국 데이터 저장 완료: 전국_대피소_데이터.csv ({len(processed_nationwide)}개)")

# 대구 데이터 저장
processed_daegu.to_csv('대구_대피소_데이터.csv', index=False, sep='|', encoding='utf-8-sig')
print(f"대구 데이터 저장 완료: 대구_대피소_데이터.csv ({len(processed_daegu)}개)")

# 요약 정보 출력
print("\n=== 데이터 요약 ===")
print(f"전국 총 대피소: {len(processed_nationwide)}개")
print(f"대구 총 대피소: {len(processed_daegu)}개")
print(f"대구 비율: {len(processed_daegu)/len(processed_nationwide)*100:.1f}%")

print("\n=== 전국 시설구분 통계 ===")
print(processed_nationwide['시설구분코드'].value_counts())

print("\n=== 대구 시설구분 통계 ===")
print(processed_daegu['시설구분코드'].value_counts())
```

# DB 스키마 매핑 분석

## API 데이터 vs DB 스키마 (shelters 테이블) 매핑 분석

### 직접 매핑 가능한 필드들 (API → DB)

| DB 필드 | API 필드 | 변환 방법 | 상태 |
|---------|----------|-----------|------|
| latitude | 위도 (계산된 값) | 도분초→십진도 변환 완료 | ✅ 완료 |
| longitude | 경도 (계산된 값) | 도분초→십진도 변환 완료 | ✅ 완료 |
| shelter_name | 시설명 (FCLT_NM) | 직접 매핑 | ✅ 완료 |
| address | 시설주소도로명 (FCLT_ADDR_RONA) | 직접 매핑 | ✅ 완료 |
| contact_phone | 관리기관전화번호 (MNG_INST_TELNO) | 직접 매핑 | ✅ 완료 |
| total_capacity | 대피가능인원수 (SHNT_PSBLTY_NOPE) | 문자열→정수 변환 필요 | ⚠️ 변환 필요 |

### 부분 매핑/변환 필요 필드들

| DB 필드 | API 필드 | 변환 방법 | 상태 |
|---------|----------|-----------|------|
| disaster_type | 시설구분코드 (FCLT_SE_CD) | 한파/무더위/지진/지진해일 → 재난유형 매핑 | ⚠️ 매핑룰 필요 |
| status | 개방여부 (OPN_YN) | Y/N → 운영중/폐쇄 매핑 | ⚠️ 매핑룰 필요 |

### 생성/가정이 필요한 필드들 (DB에 필요하지만 API에 없음)

| DB 필드 | 기본값/생성 방법 | 우선순위 |
|---------|------------------|----------|
| shelter_id | "SH_" + 시설코드 또는 자동 생성 | 🔴 필수 |
| manager_id | 기본값 또는 별도 관리자 할당 | 🔴 필수 |
| current_occupancy | 기본값 0 또는 랜덤 생성 | 🟡 중간 |
| occupancy_rate | current_occupancy/total_capacity | 🟡 중간 |
| has_disabled_facility | 기본값 false 또는 시설명 기반 추론 | 🟢 낮음 |
| has_pet_zone | 기본값 false | 🟢 낮음 |
| amenities | 빈 문자열 또는 시설구분 기반 추론 | 🟢 낮음 |
| contact_person | 빈 문자열 또는 "담당자" | 🟢 낮음 |
| contact_email | 빈 문자열 또는 기관명 기반 생성 | 🟢 낮음 |
| total_requests | 기본값 0 | 🟡 중간 |
| fulfilled_requests | 기본값 0 | 🟡 중간 |
| pending_requests | 기본값 0 | 🟡 중간 |
| created_at | 현재 시간 | 🔴 필수 |
| updated_at | 현재 시간 | 🔴 필수 |

### 사용하지 않는 API 필드들

| API 필드 | 사유 |
|----------|------|
| 시설코드 (FCLT_CD) | shelter_id 생성에만 활용 |
| 시설규모 (FCLT_SCL) | total_capacity가 더 구체적 |
| 관리기관명 (MNG_INST_NM) | 필터링에만 사용 |
| 지상지하구분 (GRND_UDGD_SE) | DB 스키마에 해당 필드 없음 |

```python
import uuid
from datetime import datetime
import random

def map_to_shelter_schema(df_processed, data_name=""):
    """API 데이터를 DB shelters 테이블 스키마에 맞게 변환"""
    
    # 기본 복사본 생성
    df_mapped = df_processed.copy()
    
    # 1. 직접 매핑 가능한 필드들
    df_mapped['shelter_name'] = df_mapped['시설명']
    df_mapped['address'] = df_mapped['시설주소도로명'] 
    df_mapped['contact_phone'] = df_mapped['관리기관전화번호']
    df_mapped['latitude'] = df_mapped['위도']
    df_mapped['longitude'] = df_mapped['경도']
    
    # 2. 변환이 필요한 필드들
    # total_capacity: 문자열 → 정수 변환
    df_mapped['total_capacity'] = pd.to_numeric(df_mapped['대피가능인원수'], errors='coerce').fillna(100).astype(int)
    
    # disaster_type: 시설구분코드 매핑
    disaster_type_mapping = {
        '한파': '한파',
        '무더위': '폭염', 
        '지진': '지진',
        '지진해일': '해일'
    }
    df_mapped['disaster_type'] = df_mapped['시설구분코드'].map(disaster_type_mapping).fillna('기타')
    
    # status: 개방여부 매핑
    status_mapping = {'Y': '운영중', 'N': '폐쇄'}
    df_mapped['status'] = df_mapped['개방여부'].map(status_mapping).fillna('운영중')
    
    # 3. 생성/가정이 필요한 필드들
    
    # 🔴 필수 필드들
    df_mapped['shelter_id'] = ['SH_' + str(i).zfill(6) for i in range(1, len(df_mapped) + 1)]
    df_mapped['manager_id'] = 'user_000000'  # 기본 관리자 (나중에 실제 관리자 할당)
    df_mapped['created_at'] = datetime.now().isoformat()
    df_mapped['updated_at'] = datetime.now().isoformat()
    
    # 🟡 중간 우선순위 필드들
    df_mapped['current_occupancy'] = 0  # 초기값 0
    df_mapped['occupancy_rate'] = 0 # 초기값 0
    df_mapped['total_requests'] = 0
    df_mapped['fulfilled_requests'] = 0  
    df_mapped['pending_requests'] = 0
    
    # 🟢 낮은 우선순위 필드들 (기본값)
    df_mapped['has_disabled_facility'] = False
    df_mapped['has_pet_zone'] = False
    df_mapped['amenities'] = ''
    df_mapped['contact_person'] = '최고관리자'
    df_mapped['contact_email'] = ''
    
    # 4. DB 스키마 순서에 맞게 컬럼 선택 및 정렬
    schema_columns = [
        'shelter_id', 'manager_id', 'shelter_name', 'disaster_type', 'status', 
        'address', 'latitude', 'longitude', 'total_capacity', 'current_occupancy', 
        'occupancy_rate', 'has_disabled_facility', 'has_pet_zone', 'amenities',
        'contact_person', 'contact_phone', 'contact_email', 'total_requests',
        'fulfilled_requests', 'pending_requests', 'created_at', 'updated_at'
    ]
    
    result = df_mapped[schema_columns]
    
    print(f"{data_name} 스키마 매핑 완료: {len(result)}개")
    print("매핑 결과 샘플:")
    print(result[['shelter_id', 'shelter_name', 'disaster_type', 'total_capacity', 'status']].head())
    
    return result

# 전국 데이터를 DB 스키마에 맞게 변환
shelter_schema_nationwide = map_to_shelter_schema(processed_nationwide, "전국")

# 대구 데이터를 DB 스키마에 맞게 변환
shelter_schema_daegu = map_to_shelter_schema(processed_daegu, "대구")
```

```python
# 스키마 매핑 결과 검증 및 저장

print("=== 스키마 매핑 검증 ===")

# 1. 데이터 타입 검증
print("\n1. 데이터 타입 검증:")
print(f"shelter_id: {shelter_schema_nationwide['shelter_id'].dtype}")
print(f"total_capacity: {shelter_schema_nationwide['total_capacity'].dtype}")
print(f"latitude: {shelter_schema_nationwide['latitude'].dtype}")
print(f"longitude: {shelter_schema_nationwide['longitude'].dtype}")

# 2. 필수 필드 null 체크
print("\n2. 필수 필드 null 체크:")
required_fields = ['shelter_id', 'manager_id', 'shelter_name', 'disaster_type', 'status']
for field in required_fields:
    null_count = shelter_schema_nationwide[field].isnull().sum()
    print(f"{field}: {null_count}개 null값")

# 3. 값 분포 확인
print("\n3. 주요 필드 값 분포:")
print("재난 유형 분포:")
print(shelter_schema_nationwide['disaster_type'].value_counts())
print("\n운영 상태 분포:")
print(shelter_schema_nationwide['status'].value_counts())

# 4. DB 스키마 형태로 저장
shelter_schema_nationwide.to_csv('shelter_schema_전국.csv', index=False, encoding='utf-8-sig')
shelter_schema_daegu.to_csv('shelter_schema_대구.csv', index=False, encoding='utf-8-sig')

print(f"\n=== 저장 완료 ===")
print(f"전국 DB 스키마 데이터: shelter_schema_전국.csv ({len(shelter_schema_nationwide)}개)")
print(f"대구 DB 스키마 데이터: shelter_schema_대구.csv ({len(shelter_schema_daegu)}개)")

# 5. 추가 처리가 필요한 항목들 요약
print("\n=== 🔴 추가 처리 필요 항목들 ===")
print("1. manager_id: 현재 'admin_001'로 통일 → 실제 관리자 할당 필요")
print("2. amenities: 현재 빈 문자열 → 시설 정보 기반 추론 로직 필요")
print("3. contact_email: 현재 빈 문자열 → 관리기관명 기반 생성 또는 수집 필요")
print("4. has_disabled_facility/has_pet_zone: 시설명 키워드 분석으로 개선 가능")

# 6. 샘플 데이터 출력
print("\n=== 최종 결과 샘플 ===")
shelter_schema_nationwide.head()
```
